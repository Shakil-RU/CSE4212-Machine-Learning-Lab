### Import Necessary Modules

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram
from sklearn.cluster import AgglomerativeClustering
```

### Generate and Plot the dataset

```python
X, y = make_blobs(n_samples=300,centers=4,random_state=42)

print(f"shape of the x: {X.shape}")
print(f"shape of the y: {y.shape}")

# Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Plot the dataset
plt.figure(figsize=(7, 7))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, cmap='viridis', s=50, edgecolor='k')
plt.title('Original Blobs Data')
plt.xlabel('Feature 1 (scaled)')
plt.ylabel('Feature 2 (scaled)')
plt.show()
```

### Plot dendrogram

```python
# Hierarchical clustering (ward linkage)
Z = linkage(X_scaled, method='ward')

# Plot dendrogram
plt.figure(figsize=(12, 6))
plt.title('Hierarchical Clustering Dendrogram (Ward linkage)')
plt.xlabel('Sample index')
plt.ylabel('Distance')
dendrogram(Z, truncate_mode='level', p=5)  
plt.show()
```

### Agglomerative Clustering Method

```python
# Agglomerative Clustering with 4 clusters
agglo = AgglomerativeClustering(n_clusters=4, metric='euclidean', linkage='ward')
agglo_labels = agglo.fit_predict(X_scaled)

plt.figure(figsize=(8, 6))

plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=agglo_labels, cmap='viridis', s=50, edgecolor='k')
plt.title('Agglomerative Clustering Result')
plt.xlabel('Feature 1 (scaled)')
plt.ylabel('Feature 2 (scaled)')

plt.tight_layout()
plt.show()
```

### Comparison between k-means and hierarchical clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

# Generate two moons dataset
X, y_true = make_moons(n_samples=300, noise=0.05, random_state=42)

# Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# K-Means clustering (k=2)
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Agglomerative Clustering (ward linkage, k=2)
agglo = AgglomerativeClustering(n_clusters=2, linkage='ward')
agglo_labels = agglo.fit_predict(X_scaled)

# Plot using plt.subplot
plt.figure(figsize=(18, 5))

# Original data with true labels
plt.subplot(1, 3, 1)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_true, cmap='viridis', s=50, edgecolor='k')
plt.title('Original Data(Original)')
plt.xlabel('Feature 1 (scaled)')
plt.ylabel('Feature 2 (scaled)')

# K-Means clustering result
plt.subplot(1, 3, 2)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels, cmap='viridis', s=50, edgecolor='k')
plt.title('K-Means Clustering Result')
plt.xlabel('Feature 1 (scaled)')
plt.ylabel('Feature 2 (scaled)')

# Agglomerative Clustering result
plt.subplot(1, 3, 3)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=agglo_labels, cmap='viridis', s=50, edgecolor='k')
plt.title('Agglomerative Clustering Result')
plt.xlabel('Feature 1 (scaled)')
plt.ylabel('Feature 2 (scaled)')

plt.tight_layout()
plt.show()
```
